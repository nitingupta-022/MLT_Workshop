# Sampling from Distributions, Bar Plots, Histograms And Scatter Plots

## Import & Settings

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(seed = 1)
```

## Sampling And Plotting

NumPy provides a utility to generate pseudo-random numbers. We shall try to sample points from various distributions. Once we have a sample, we can then represent it pictorially suitable plots.

### Sampling : Bernoulli

Let us generate a sample of 1000 points from the Br(0.7)

```python
X = rng.choice([0,1], p = [0.3, 0.7], size = 10)
X
```

```python
# The outcomes are 0 and 1.
# We want to assign the probability of success to be 0.7.

X = rng.choice([0,1], p = [0.3, 0.7], size = 1000)
X.shape
```

### Plotting : Bar Plot

Let us visualize the sample using a bar plot. For this, we first need the height of the bars.

```python
zero, one = 0, 0
for i in range(X.shape[0]) :
  if X[i] == 0 :
    zero += 1
  else :
    one += 1

print(zero, one)
```

```python
plt.bar([0,1], [zero, one], width = 0.5)
plt.xticks([0,1])
```

### Sampling : Bivariate Gaussian

Sample 1000 points from the following Bivariate Gaussian :    

$$
\mathcal{N} \left(\begin{bmatrix}1 \\ 2\end{bmatrix}, \begin{bmatrix}1 & 0\\0 & 5\end{bmatrix} \right)
$$

```python
mu = np.array([1,2])
cov = np.array([[1,0],
                [0,5]])

X = rng.multivariate_normal(mu, cov, size = 100000).T
# We transpose so that the data-matrix is (d,n) and not (n,d)
d,n = X.shape
```

```python
plt.scatter(X[0 , :], X[1, : ])
```

### Plotting : Scatter Plot

Let us now visualise the above 1,00,000 sample using a scatter plot. Zooming out of the scatter plot gives us a better understanding. Changing the covariance matrix also helps in understanding how the sampled points depend on it.

```python
plt.scatter(X[0 , : ], X[1, : ])
plt.axis('equal') # this ensures the equal scaling on both the axes
plt.axhline(color = 'black', linestyle = '--', linewidth = '0.8')
plt.axvline(color = 'black', linestyle = '--', linewidth = '0.8')
```

### Estimating the sample covariance matrix

Let us now estimate the sample covariance matrix and verify it it is close to the population covariance matrix. For this, we use extend the concept of a norm to matrices.

```python
# Covariance matrix : (XX^T)/n, when X is centered
d,n = X.shape
mu = X.mean(axis = 1).reshape(d,1)
print(mu)
C = ((X - mu) @ (X - mu).T)/n
C # Sample covariance
```

```python
# Check how close C and cov are

np.linalg.norm(C - cov)
```

Let us now draw 1,000,000 samples from a Gaussian Mixture Model (GMM) that has three components, with mixture probabilities **[0.2, 0.3, 0.5]** and means **[0, 5, 10]**. The standard deviation of all three Gaussians is the same and is equal to 1. We can visualise the sample using a histogram

- First choose the component
- Generate a point from the chosen component

```python
# For 1_00_000 samples

n = 1_00_000
comp = rng.choice([0,1,2], p = [0.2, 0.3, 0.5], size = n)
comp[ : 50]
```

```python
n = 1_00_000
comp = rng.choice([0,1,2], p = [0.2, 0.3, 0.5], size = n)
mu = np.array([0, 5, 10])

X = np.zeros(n)
for i in range(n) :
  X[i] = rng.normal(mu[comp[i]], 1)

X.shape
```

```python
plt.hist(X, bins = 1000, edgecolor = 'black')
```
