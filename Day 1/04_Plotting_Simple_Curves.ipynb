### Curve - 2

Plot ***y*** = 5 - 3***x*** ùõú **[-5, 5]**. Sample 20 equally spaced points in this interval, inclusive of endpoints. Add a title to the plot which has the equation of the curve.

### Curve - 3

Plot ***y*** = ***x^2*** for ***x*** ùûä **[-1, 1]**. Try out four different samples for **x**
- 5 equally spaced points in this interval, endpoints inclusive
- 10 equally spaced points in this interval, endpoints inclusive
- 20 equally spaced points in this interval, endpoints inclusive
- 50 equally spaced points in this interval, endpoints inclusive

Observe the differences in these four plots. Which one would we choose ? Add the equation of the curve as the title of the plot. Also add the x and y-axis to the plot

```python
# For 5 points
x = np.linspace(-1, 1, 5)
y = x**2

plt.scatter(x,y)
plt.plot(x,y)
plt.axhline(color = 'black', linestyle = '--', linewidth = 0.8)
plt.axvline(color = 'black', linestyle = '--', linewidth = 0.8)
```

```python
# For 10 points
x = np.linspace(-1, 1, 10)
y = x**2

plt.scatter(x,y)
plt.plot(x,y)
plt.axhline(color = 'black', linestyle = '--', linewidth = 0.8)
plt.axvline(color = 'black', linestyle = '--', linewidth = 0.8)
```

```python
# For 20 points
x = np.linspace(-1, 1, 50)
y = x**2

plt.scatter(x, y)
plt.plot(x,y)
plt.axhline(color = 'black', linestyle = '--', linewidth = 0.8)
plt.axvline(color = 'black', linestyle = '--', linewidth = 0.8)
```

### Curve - 4

Plot ***y*** = 3 and ***x*** = 5 on the same plot. Color the first one green and the second one red. Limit your plot to the region ***x*** ùùê **[0,6]** and ***y*** ùùê **[0,6]**

- Label the x and y axes.
- Add a suitalbe title
- Add a legend

```python
# y = 3
x = np.linspace(0,6, 2)
y = np.ones(2) * 3
plt.plot(x,y, color = 'green', label = 'y = 3')

# x = 5
y = np.linspace(0,6,2)
x = np.ones(2) * 5
plt.plot(x, y, color = 'red', label = 'x = 5')

plt.legend()
plt.title("x = 5 & y = 3")
plt.axhline(color = 'black', linestyle = '--')
plt.axvline(color = 'black', linestyle = '--')
```

## Subplots

Sometimes we may have to plot multiple objects, each in a separate plot. Plot the following curves, each in a separate plot :     
- y = **ln x**, ***x*** ùûä (1/e^2 , e^3)
- y = **e^x** , ***x*** ùûä (0,3)
- y = **sin x** , ***x*** ùûä [0, 2œÄ]
- y = 1/x ,  ***x*** ùûä (0, 5)

Title each curve appropriately

```python
# rcParams is like a dict
# controls various configurations
plt.rcParams['figure.figsize'] = [8,8]
plt.rcParams['font.size'] = 10

# Plot - 1
plt.subplot(2,2,1)
x = np.linspace(1/np.e**2, np.e**3)
y = np.log(x)
plt.plot(x, y)
plt.title('y = ln x')

# Plot - 2
plt.subplot(2,2,2)
x = np.linspace(0,3)
y = np.exp(x)
plt.plot(x, y)
plt.title('y = e^x')

# Plot - 3
plt.subplot(2,2,3)
x = np.linspace(0, 2 * np.pi)
y = np.sin(x)
plt.plot(x,y)
plt.title('y = sin x')

# Plot - 4
plt.subplot(2,2,4)
x = np.linspace(0,5)
y = 1/x
plt.plot(x, y)
plt.title('y = 1/x')
```
